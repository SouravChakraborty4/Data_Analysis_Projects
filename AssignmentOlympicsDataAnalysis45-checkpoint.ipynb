```python
import pandas as pd
```

```python
DatasetAthlete_events = pd.read_csv("athlete_events.csv")
DatasetAthlete_events
```

```python
DatasetRegions = pd.read_csv("regions1.csv")
DatasetRegions
```

```python
#joining the data based on appropriate column.
```

```python
dataset1 = pd.merge(DatasetAthlete_events, DatasetRegions, on='NOC', how='left')
dataset1
```

```python
#Replacing nan values with the columnâ€™s mean, For numeric columns
```

```python
import pandas as pd
import numpy as np
```

```python
dataset1['Age'] = dataset1['Age'].fillna(dataset1['Age'].mean())
dataset1['Height'] = dataset1['Height'].fillna(dataset1['Height'].mean())
dataset1['Weight'] = dataset1['Weight'].fillna(dataset1['Weight'].mean())
dataset1
```

```python
#Top 10 countries with most gold, silver, bronze medal
```

```python
goldMedals = dataset1[(dataset1.Medal == 'Gold')]
goldMedals
goldMedals.region.value_counts().reset_index(name='Medal').head(10)
```

```python
SilverMedals = dataset1[(dataset1.Medal == 'Silver')]
SilverMedals
SilverMedals.region.value_counts().reset_index(name='Medal').head(10)
```

```python
BronzeMedals = dataset1[(dataset1.Medal == 'Bronze')]
BronzeMedals
BronzeMedals.region.value_counts().reset_index(name='Medal').head(10)
```

```python
#Finding the distribution of gold medalists according to their age.
```
